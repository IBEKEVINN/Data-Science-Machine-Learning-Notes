# FRAMEWORK

**Markov Decision Process**
- Only requires current state, not history of all states and actions

![Alt text](image.png)

**Exploration / Exploitation Trade-off**
* Exploration - Trying **random actions** to find out more information about the environment.
* Exploitation - Using information to** maximise reward**.

**Policy-Based Method**

* The policy ($\pi$) is the function we want to learn, which maximises expected return. We find it through training.
* Could learn either which __action to take__ or which __state is better__.<br>
--
* **Deterministic** - A definitive mapping for each state to best corresponding action.
* **Schotastic** - A probability distribution to possible actions.

**Value-Based Method**

* Given a state, our action-value function (that we train) outputs the value of each action at that state. Then, our pre-defined Greedy Policy selects the action that will yield the highest value given a state or a state action pair.
* Maps an expected value to each state
* Value is the _expected_ _discounted_ return the agent gets if it starts in that state, then acts according to current policy.


**Types of Tasks**
* Episodic: Starting point and an ending point
* Continuing: Task that continue forever (no terminal state)
